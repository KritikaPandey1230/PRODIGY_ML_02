```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
customer_data = pd.read_csv("Mall_Customers.csv")
```

```python
print(customer_data.head())
```

```python
print("Shape of the dataset:", customer_data.shape)
```

```python
print("Dataset Information:")
print(customer_data.info())
```

```python
print("Missing values in each column:")
print(customer_data.isnull().sum())
```

```python
# Selecting relevant features for clustering
X = customer_data.iloc[:, [3, 4]].values
print("Feature values for clustering:\n", X)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the Elbow graph
sns.set()
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))  # Set x-ticks for clarity
plt.grid()
plt.show()
```

```python
# Fitting KMeans with the chosen number of clusters
optimal_clusters = 5  # Change this based on the Elbow method result
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=0)
Y = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0],X[Y==0,1],s=50,c='green',label='Cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50,c='red',label='Cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50,c='yellow',label='Cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50,c='orange',label='Cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50,c='blue',label='Cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='violet',label='Centroids')
plt.title('Customer Group')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()



```
